## Leetcode 896 - Monotonic Array

Kiểm tra các phần tử trong mảng có không đổi/tăng dần hoặc không đổi/giảm dần hay không.

[Link bài tập](https://leetcode.com/problems/monotonic-array/)


```
Example 1:

Input: nums = [1,2,2,3]
Output: true

Example 2:

Input: nums = [6,5,4,4]
Output: true

Example 3:

Input: nums = [1,3,2]
Output: false
```

````python
def isMonotonic(nums: List[int]) -> bool:
    # Lập trình tại đây
    increasing = decreasing = True

    for i in range(len(nums) - 1):
        if nums[i] > nums[i+1]:
            increasing = False
        if nums[i] < nums[i+1]:
            decreasing = False

    return increasing or decreasing
````
