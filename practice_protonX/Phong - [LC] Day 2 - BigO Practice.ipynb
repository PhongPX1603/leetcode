## 1295. Find Numbers with Even Number of Digits

Tìm số lượng các số có số lượng chẵn ký tự

[Link bài tập](https://leetcode.com/problems/find-numbers-with-even-number-of-digits/)

### ProtonX Leetcode 200 - Day 2 - Homework
Cách lập trình:
- Bạn lập trình thành công trên Leetcode
- Sau đó bạn thân hàm vào trong những phần có ghi chú **Lập trình tại đây**
- Sau đó bạn nộp lại bài tập qua hệ thống ProtonX để tính điểm

Một câu: 60 điểm

````python
from typing import List
````

**Example 1:**

```
Input: nums = [12,345,2,6,7896]
Output: 2
```

```
Explanation: 
12 contains 2 digits (even number of digits). 
345 contains 3 digits (odd number of digits). 
2 contains 1 digit (odd number of digits). 
6 contains 1 digit (odd number of digits). 
7896 contains 4 digits (even number of digits). 
Therefore only 12 and 7896 contain an even number of digits.

```

**Example 2:**

```
Input: nums = [555,901,482,1771]
Output: 1 
Explanation: 
Only 1771 contains an even number of digits.

```

**Constraints:**
```
1 <= nums.length <= 500
1 <= nums[i] <= 105
```

````python
def findNumbers(nums: List[int]) -> int:
    # Lập trình tại đây
    result = 0
    for num in nums:
        if len(str(num)) % 2 == 0:
            result += 1
    return result
````

## 1. Two Sum

Tìm chỉ mục của 2 phần tử có tổng bằng một số cho trước

[Link bài tập](https://leetcode.com/problems/two-sum/)


**Example 1:**

```
Input: nums = [2,7,11,15], target = 9
Output: [0,1]
Explanation: Because nums[0] + nums[1] == 9, we return [0, 1].
```

**Example 2:**
```
Input: nums = [3,2,4], target = 6
Output: [1,2]
```
**Example 3:**
```
Input: nums = [3,3], target = 6
Output: [0,1]
```

**Constraints:**
```
2 <= nums.length <= 104
-109 <= nums[i] <= 109
-109 <= target <= 109

```
**Only one valid answer exists.**

````python
def twoSum(nums: List[int], target: int) -> List[int]:
    # Lập trình tại đây
    result = []
    for num in nums:
        if target - num in nums[nums.index(num)+1:]:
            result.append(nums.index(num))
            second = nums[nums.index(num)+1:].index(target - num) + nums.index(num) + 1
            result.append(second)
            break
    return result
````
