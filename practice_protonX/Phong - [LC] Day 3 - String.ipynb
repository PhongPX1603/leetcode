## 2. Leetcode 345 - Reverse Vowels of a String

Đảo ngược nguyên âm của một chuỗi

[Link bài tập](https://leetcode.com/problems/reverse-vowels-of-a-string/)


**Example 1:**

```
Input: s = "hello"
Output: "holle"
```

**Example 2:**
```
Input: s = "leetcode"
Output: "leotcede"
```


**Constraints:**
```
1 <= s.length <= 3 * 105
s consist of printable ASCII characters.

```

````python
def reverseVowels(s: str) -> str:
    # Lập trình tại đây
    result = list(s)
    i = 0
    j = len(s) - 1
    while i<=j:
        if result[i] in ['u', 'e', 'o', 'a', 'i'] and result[j] in ['u', 'e', 'o', 'a', 'i']:
            result[i], result[j] = result[j], result[i]
            i += 1
            j -= 1
        elif result[i] in ['u', 'e', 'o', 'a', 'i'] and result[j] not in ['u', 'e', 'o', 'a', 'i']:
            j -= 1
        elif result[i] not in ['u', 'e', 'o', 'a', 'i'] and result[j] in ['u', 'e', 'o', 'a', 'i']:
            i += 1
        else:
            i += 1
            j -= 1
    result = ''.join(result)

    return result
````

## 3. Đếm số lượng chuỗi con chỉ có duy nhất một loại chữ



**Ràng buộc:**
```
1 <= s.length <= 1000
Chuỗi chỉ chứa chữ cái tiếng Anh viết thường.
```

**Example 1:**

```
Input: s = "aaaba"
Output: 8

Giải thích: Các chuỗi con với duy nhất 1 loại chữ đó là "aaa", "aa", "a", "b" và
- "aaa" xuất hiện 1 lần
- "aa" xuất hiện 2 lần
- "a" xuất hiện 4 lần
- "b" xuất hiện 1 lần

Tổng số lần xuất hiện là 1 + 2 + 4 + 1 = 8 lần
```

````python
def countSubStringTimes(s: str) -> int:
    # Lập trình tại đây
    result = 0
    count = 1
    for i in range(len(s)-1):
        if s[i] == s[i+1]:
            count += 1
            print('count', count)
            if i == len(s) - 2:
                k = 0
                for j in range(1, count+1):
                    k += j
                    print('k', k)
                result += k

        else:
            k = 0
            for j in range(1, count+1):
                k += j
                print('k', k)
            result += k
            count = 1
            print('result', result)
    if s[-1] != s[-2]:
        result += 1
    return result
````

## 4. Leetcode 541 - Đảo ngược chuỗi 2

Đảo ngược k kí tự trong 2k ký tự bắt đầu của chuỗi

[Link bài tập](https://leetcode.com/problems/reverse-string-ii/)



**Example 1:**

```
Input: s = "abcdefg", k = 2
Output: "bacdfeg"
```

**Example 2:**

```
Input: s = "abcd", k = 2
Output: "bacd"
```

**Constraints:**

```
1 <= s.length <= 104
s consists of only lowercase English letters.
1 <= k <= 104
```

````python
def reverseStr(s: str, k: int) -> str:
    # Lập trình tại đây
    result = list(s)
    if k >= len(s):
        result.reverse()
        return ''.join(result)
    for i in range(0, len(s), 2 * k):
        result[i:i + k] = reversed(result[i:i + k])
    result = ''.join(result)
    return result
````

## 5. Leetcode - 1347. Minimum Number of Steps to Make Two Strings Anagram

Số bước nhỏ nhất để tạo ra 2 chuỗi Anagram.

[Link bài tập](https://leetcode.com/problems/minimum-number-of-steps-to-make-two-strings-anagram/)

**Example 1:**

```
Input: s = "bab", t = "aba"
Output: 1
Explanation: Replace the first 'a' in t with b, t = "bba" which is anagram of s.
```

**Example 2:**

```
Input: s = "leetcode", t = "practice"
Output: 5
Explanation: Replace 'p', 'r', 'a', 'i' and 'c' from t with proper characters to make t anagram of s.
```
**Example 3:**

```
Input: s = "anagram", t = "mangaar"
Output: 0
Explanation: "anagram" and "mangaar" are anagrams. 
```

**Constraints:**

```
1 <= s.length <= 5 * 104
s.length == t.length
s and t consist of lowercase English letters only.
```

````python
def minSteps(s: str, t: str) -> int:
    result = 0
    import collections
    counter_s = collections.Counter(s)
    counter_t = collections.Counter(t)
    for key in counter_s.keys():
        if key in counter_t.keys():
            if counter_s[key] > counter_t[key]:
                result = result + (counter_s[key] - counter_t[key])
        else:
            result += counter_s[key]

    return result
````

## 6. Leetcode 125 - Valid Palindrome

Làm sạch chuỗi và kiểm tra chuỗi Palindrome.

[Link bài tập](https://leetcode.com/problems/valid-palindrome/)

**Example 1:**

```
Input: s = "A man, a plan, a canal: Panama"
Output: true
Explanation: "amanaplanacanalpanama" is a palindrome.
```



**Example 2:**

```
Input: s = "race a car"
Output: false
Explanation: "raceacar" is not a palindrome.
```

**Example 3:**
```
Input: s = " "
Output: true
Explanation: s is an empty string "" after removing non-alphanumeric characters.
Since an empty string reads the same forward and backward, it is a palindrome.
```

**Constraints:**

```
1 <= s.length <= 2 * 105
s consists only of printable ASCII characters.
```

````python
def isPalindrome(s: str) -> bool:
    result = True
    i = 0
    j = len(s) - 1
    while i<=j:
        if s[i].isalnum() and s[j].isalnum():
            if s[i].lower() != s[j].lower():
                result = False
                break
            else:
                i += 1
                j -= 1
        else:
            if s[i].isalnum() == False:
                i += 1
            if s[j].isalnum() == False:
                j -= 1
    return result
````
